## Inspect observation query results from `astroquery.mast.MastMissions`

```python
import os
from astropy.table import Table
from astropy.coordinates import SkyCoord

from astroquery.mast import Observations, MastMissions
from mast_table import CrossFilterMastTable

mm = MastMissions(mission='roman')

# use mast TEST
mm._service_api_connection.SERVICE_URL = 'https://masttest.stsci.edu'
mm._service_api_connection.REQUEST_URL = 'https://masttest.stsci.edu/search/roman/api/v0.1/'
mm._service_api_connection.MISSIONS_DOWNLOAD_URL = 'https://masttest.stsci.edu/search/'

coordinates = SkyCoord(ra=80, dec=-70, unit='deg')

cached_query_path = 'cached_query.ecsv'
if os.path.exists(cached_query_path):
    observations = Table.read(cached_query_path)
else:
    observations = mm.query_region(
        coordinates=coordinates,
        radius='10 arcmin',
        program=164,
        exposure_type='WFI_IMAGE'
    )
    observations.write(cached_query_path, overwrite=True)

mast_table_1 = CrossFilterMastTable(observations)
mast_table_1
```

## Inspect query results from `astroquery.mast.Catalogs`

```python
from astroquery.mast import Catalogs
from astropy.coordinates import SkyCoord
target = SkyCoord.from_name('TRAPPIST-1')

catalog_data = Catalogs.query_region(target,
                                     radius=0.01,
                                     catalog="Panstarrs",
                                     data_release="dr1",
                                     table="mean")

mast_table_2 = CrossFilterMastTable(catalog_data)
mast_table_2
```

## Inspect catalog query results from `astroquery.mast.Observations`

```python
observations = Observations.query_criteria(object_name="Kepler-444", radius=0.01, obs_collection='GALEX')

mast_table_3 = CrossFilterMastTable(observations)
mast_table_3
```
